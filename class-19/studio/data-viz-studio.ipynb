```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to 
## get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", on_bad_lines='skip')

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

df.head()
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.
#  plotting the total number of books published per year
published_year = np.sort(df["year"].unique())
number_books_by_year = df.groupby("year")["bookID"].count()

plt.plot(published_year, number_books_by_year)
plt.xlabel("Year")
plt.ylabel("Number of Books")
plt.title("Books Published per Year")
plt.xticks(rotation=45)
plt.show()
```

```python
# Use your Python skills to create a visualization here.
# since the above visualization is not vey clear, will try to create to a visualization where the years are grouped in bins 
# and each bin is of duration 10 years and plotting the total books published during that time frame. 

# Unique year
published_year = np.sort(df["year"].unique())

# count of books per year 
books_per_year = df.groupby("year")["bookID"].count()

# Convert 'year' to integer
df['year'] = df['year'].astype(int)

# Create 20-year bins
bins = np.arange(df["year"].min(), df["year"].max() + 10, 10)

# Aggregate counts into 20-year bins
books_per_bin = books_per_year.groupby(pd.cut(books_per_year.index, bins)).sum()

# Plot as bar chart
figsize=(10,8)
books_per_bin.plot(kind='bar', color = "orange", edgecolor = "black")
plt.xlabel("Year Range")
plt.ylabel("Number of Books")
plt.title("Number of Books Published per 20-Year Bin")
plt.xticks(rotation=45)
plt.show()
```
